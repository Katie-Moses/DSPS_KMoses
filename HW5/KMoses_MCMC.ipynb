```python
import numpy as np
import pylab as plt
#!pip install astropy
import astropy
from astropy import units as u
from scipy.stats import norm

np.random.seed(661)
```

```python
steps = 10000
x_vals = np.zeros(steps)
x_initializer = 0

i = 0
while(i<steps):
  # random x sampler
  x_sampler = np.random.normal(x_initializer,1)

  # Gaussian Density w/ mean 2 and variance 2
  alpha_level = norm.pdf(x_sampler, loc=2, scale=np.sqrt(2)) / norm.pdf(x_initializer, loc=2, scale=np.sqrt(2))

  # MCMC thing
  if alpha_level > np.random.rand():
    x_initializer = x_sampler

  x_vals[i] = x_sampler
  i+=1

support = np.arange(-10,15,.25)
norm_2 = norm.pdf(support, loc=2,scale=np.sqrt(2))
plt.hist(x_vals, bins=50, density=True);
plt.plot(support,norm_2)
plt.xlabel("x")
plt.ylabel("p(x)")
plt.figtext(0.5, -0.05, "Figure 1: Simple M-H MCMC Sampler based on the constraints in Problem 2 of Hogg and Foreman-Mackey 2017.", ha='center', fontsize=10, wrap=True)
plt.title("M-H MCMC Sampler")
```
